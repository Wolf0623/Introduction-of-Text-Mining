```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re
import nltk
import pandas as pd
```

```python
# 讀取txt檔並存進doc內
docs = []
# 特殊符號的清單
remove_chars = '[0-9’!"#$%&\'()*+,-./:;<=>?@，。?★、…【】《》？“”‘’！[\\]^_`{|}~]+'

for i in range(1, 1096):
    raw = open("C:\\Users\\HP\\Desktop\\文字探勘初論\\PA1-data\\" + str(i) + ".txt","r")
    temp = raw.read()

    # 移除換行符號且換小寫
    temp_lowered = temp.replace("\n", "").replace("\r", "").lower()

    # 移除標點符號
    temp_cleared = re.sub(remove_chars, "", temp_lowered)
    
    # 存進doc
    docs.append(temp_cleared)
```

```python
# 參考上課投影片使用的停止字
nltk.download("stopwords")
stop_words = nltk.corpus.stopwords.words("english")
len(stop_words)
```

```python
stop_words
```

```python
# 把要用的stop_word加進去
TF_vectorizer = CountVectorizer(stop_words=stop_words)
```

```python
TF_vectors = TF_vectorizer.fit_transform(docs)
```

```python
TF_vectorizer.get_feature_names()
```

```python
print(TF_vectors.toarray())
```

```python
binary_vectorizer = CountVectorizer(binary=True)
binary_vectors = binary_vectorizer.fit_transform(docs)
print(binary_vectors.toarray())
```

```python
# 用參數加入停止字
TFIDF_vectorizer = TfidfVectorizer(stop_words="english", lowercase=True)
TFIDF_vectors = TFIDF_vectorizer.fit_transform(docs)
# 把 feature name 存起來 
vec_name = TFIDF_vectorizer.get_feature_names()
print(TFIDF_vectors.toarray())
```

```python
# 根據題目要求把 documents存起來(每個都是20194維度的向量)
for j in range(10):
    with open('vec' + str(j+1) + '.txt', 'w') as f:
        for i in range(len(vec_name)):
            # 把 vec_name 跟對應的tf-idf值存在同一行
            f.writelines(" ".join([vec_name[i], str(TFIDF_vectors[j].toarray()[0][i])]))
            f.write('\n')  #换行
```

```python
# 讀取 vec1, vec2 算 cosine similarity
vec1 = pd.read_table('vec1.txt', header=None, encoding='utf-8', sep=' ')
vec1.columns = ["feature","tfidf"]

vec2 = pd.read_table('vec2.txt', header=None, encoding='utf-8', sep=' ')
vec2.columns = ["feature","tfidf"]
```

```python
vec1_list = vec1["tfidf"].values.tolist()
vec2_list = vec2["tfidf"].values.tolist()

cosine_similarity([vec1_list], [vec2_list])
```

```python
vec_a = vec1_list
vec_b = vec2_list

# Dot and norm
dot = sum(a*b for a, b in zip(vec_a, vec_b))
norm_a = sum(a*a for a in vec_a) ** 0.5
norm_b = sum(b*b for b in vec_b) ** 0.5

# Cosine similarity
cos_sim = dot / (norm_a*norm_b)

# Results
print('My version:', cos_sim)
```
